```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cm = np.array([[149,15],
      [6,0]])
```

```python
cm
```

```python
sns.heatmap(cm, 
			annot=True,
			fmt='g', 
			xticklabels=['Leopard','Background'],
			yticklabels=['Leopard','Background'],
             cmap="Blues")
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```
